```python
from datasets import load_dataset
import pandas as pd
import torch
import os
from torch.nn import functional as F
from transformers import AutoModelForImageClassification, TrainingArguments, Trainer
from transformers import DefaultDataCollator
from transformers import EarlyStoppingCallback
from pytorch_model_summary import summary
from datasets import Dataset
import evaluate
import transformers
transformers.set_seed(777) #하이퍼파라미터 튜닝을 위해 실행시 마다 변수가 같은 초기값 가지게 하기
import transformers_helper
transformers_helper.register_models()

dataset_dict = load_dataset("csv", data_files={"train": "custom-datasets/tabular_classification/bank_loan/train.csv", "validation": "custom-datasets/tabular_classification/bank_loan/validation.csv"})
#dataset_dict = load_dataset("json", data_files={"train": "custom-datasets/tabular_classification/bank_loan/train.json", "validation": "custom-datasets/tabular_classification/bank_loan/validation.json"})
#print(dataset)
'''
DatasetDict({
    train: Dataset({
        features: ['Loan_ID', 'Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History', 'Property_Area', 'Target'],
        num_rows: 307
    })
    val: Dataset({
        features: ['Loan_ID', 'Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History', 'Property_Area', 'Target'],
        num_rows: 77
    })
})
'''

train_dataset = dataset_dict['train']
validation_dataset = dataset_dict['validation']

#print(train_dataset)
'''
Dataset({
    features: ['Loan_ID', 'Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History', 'Property_Area', 'Target'],
    num_rows: 307
})
'''

train_df = train_dataset.to_pandas()
validation_df = validation_dataset.to_pandas()

#print(train_df.head())
'''
   Unnamed: 0   Loan_ID  Gender Married Dependents     Education  \
0           0  LP001872    Male      No          0      Graduate   
1           1  LP001222  Female      No          0      Graduate   
2           2  LP001155  Female     Yes          0  Not Graduate   
3           3  LP001529    Male     Yes          0      Graduate   
4           4  LP001255    Male      No          0      Graduate   

  Self_Employed  ApplicantIncome  CoapplicantIncome  LoanAmount  \
0           Yes             5166                0.0         128   
1            No             4166                0.0         116   
2            No             1928             1644.0         100   
3           Yes             2577             3750.0         152   
4            No             3750                0.0         113   

   Loan_Amount_Term  Credit_History Property_Area Target  
0               360               1     Semiurban      Y  
1               360               0     Semiurban      N  
2               360               1     Semiurban      Y  
3               360               1         Rural      Y  
4               480               1         Urban      N  
'''

feature_extractor = transformers_helper.models.TabularFeatureExtractor(
    continuous_columns=['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History'],
    categorical_columns=['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area'],
    labels_column='Target',
    labels_shape='one',
    labels_type='float'
)
feature_extractor.train(train_df)

labels = feature_extractor.labels
#print(labels) #['N' 'Y']
labels = ['비승인', '승인']

id2label = {}
label2id = {}
for i, label in enumerate(labels):
    id2label[i] = label
    label2id[label] = i

def transform(batch):
    #print(list(batch.keys())) #['Loan_ID', 'Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History', 'Property_Area', 'Target']
    data = {
        'Gender': batch['Gender'],
        'Married': batch['Married'],
        'Dependents': batch['Dependents'],
        'Education': batch['Education'],
        'Self_Employed': batch['Self_Employed'],
        'ApplicantIncome': batch['ApplicantIncome'],
        'CoapplicantIncome': batch['CoapplicantIncome'],
        'LoanAmount': batch['LoanAmount'],
        'Loan_Amount_Term': batch['Loan_Amount_Term'],
        'Credit_History': batch['Credit_History'],
        'Property_Area': batch['Property_Area'],
        'Target': batch['Target']
    }
    df = pd.DataFrame(data)
    inputs = feature_extractor(df, return_tensors='pt')
    batch_ = {}
    batch_['x'] = inputs['x']
    batch_['labels'] = inputs['labels']
    return batch_

train_dataset = train_dataset.map(transform, batched=True)
train_dataset = train_dataset.with_format(type='torch')
validation_dataset = validation_dataset.map(transform, batched=True)
validation_dataset = validation_dataset.with_format(type='torch')

#print(train_dataset)
'''
Dataset({
    features: ['x', 'labels'],
    num_rows: 8
})
'''

config = transformers_helper.models.AnnConfigForTabularBinaryClassification(
    in_features=20,
    return_probabilities=False,
    id2label=id2label,
    label2id=label2id
)
model = transformers_helper.models.AnnForTabularBinaryClassification(config)
#print(model)
'''
'''
#print(summary(model, torch.rand((20,)), show_input=True))
'''
'''

##########학습 루프

training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=30, #디폴트 3
    per_device_train_batch_size=16, #디폴트 8
    learning_rate=0.01, #디폴트 0.00005
    #optim='adamw_hf', #adamw_hf, adamw_torch, adamw_apex_fused, adamw_anyprecision or adafactor, 디폴트 adamw_hf #
    #lr_scheduler_type='linear', #linear, cosine, cosine_with_restarts, polynomial, constant, constant_with_warmup, 디폴트 linear #
    #weight_decay=0 #가중치 규제, 디폴트 0 #가중치 w들의 제곱 합 (L2 놈, L2 규제)을 비용 함수에 추가
    load_best_model_at_end=True, #디폴트 False #마지막에 최적 모델 로드 시작
    #metric_for_best_model='accuracy', #디폴트 loss
    #greater_is_better=True, #디폴트 True
    evaluation_strategy='epoch', #디폴트 no
    save_strategy='epoch' #디폴트 steps #마지막에 최적 모델 로드 끝
)

def compute_metrics(eval_prediction):
    validation_metrics = evaluate.combine([
        evaluate.load('accuracy')
    ])
    predictions = eval_prediction.predictions #예측 값
    label_ids = eval_prediction.label_ids #실제 값
    #print(predictions.shape) #
    #print(label_ids.shape) #
    predictions = (torch.Tensor(predictions) > 0.5).int() 
    validation_metrics_value = validation_metrics.compute(predictions=predictions, references=label_ids)
    #print(validation_metrics_value) #{'accuracy': 0.3}
    return validation_metrics_value

trainer = Trainer(
    model=model,
    args=training_args,
    #optimizers=(AdamW(model.parameters(), lr=0.001), get_linear_schedule_with_warmup(optimizer)) #디폴트 (AdamW(model.parameters(), lr=1e-3), get_linear_schedule_with_warmup(optimizer)) #
    train_dataset=train_dataset,
    eval_dataset=validation_dataset,
    #data_collator=DefaultDataCollator(), #디폴트 DefaultDataCollator() (tokenizer 옵션 지정되 있지 않으면), DataCollatorWithPadding(tokenizer=tokenizer) (tokenizer 옵션 지정되 있으면)
    #callbacks=None, #디폴트 None, 얼리스타핑 [EarlyStoppingCallback(early_stopping_patience=10)]
    compute_metrics=compute_metrics
)

trainer.train()

model_path = '/content/drive/MyDrive/models/transformers/models/bank-loan-model-for-tabular-classification'
model.save_pretrained(model_path)
feature_extractor.save_pretrained(model_path)

##########학습 과정 시각화

df = pd.DataFrame(trainer.state.log_history)
#print(df)
'''
   eval_loss  eval_accuracy  eval_runtime  eval_samples_per_second  \
0   1.144318       0.316667        0.5577                  215.182   
1   1.129987       0.316667        0.6194                  193.748   
2   1.125339       0.316667        0.5805                  206.736   
3        NaN            NaN           NaN                      NaN   

   eval_steps_per_second  epoch  step  train_runtime  \
0                 26.898    1.0    15            NaN   
1                 24.218    2.0    30            NaN   
2                 25.842    3.0    45            NaN   
3                    NaN    3.0    45         2.2495   

   train_samples_per_second  train_steps_per_second  total_flos  train_loss  
0                       NaN                     NaN         NaN         NaN  
1                       NaN                     NaN         NaN         NaN  
2                       NaN                     NaN         NaN         NaN  
3                   160.032                  20.004         0.0    1.140103 
'''

import matplotlib.pyplot as plt
plt.plot(range(1, len(df["eval_loss"]) + 1), df["eval_loss"], 'b', label='validation loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
plt.plot(range(1, len(df["eval_loss"]) + 1), df["eval_accuracy"], 'r--',label='validation accuracy')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
import torch
from transformers import AutoFeatureExtractor
import transformers_helper
transformers_helper.register_models()
import numpy as np
from transformers import pipeline
import pandas as pd

model_path = '/content/drive/MyDrive/models/transformers/models/bank-loan-model-for-tabular-classification'
model = transformers_helper.models.AnnForTabularBinaryClassification.from_pretrained(model_path)
feature_extractor = AutoFeatureExtractor.from_pretrained(model_path)
p = pipeline(task='tabular-binary-classification', model=model, feature_extractor=feature_extractor)

data = {
    'Gender': ['Male'],
    'Married': ['Yes'],
    'Dependents': ['0'],
    'Education': ['Graduate'],
    'Self_Employed': ['No'],
    'ApplicantIncome': [14583],
    'CoapplicantIncome': [0],
    'LoanAmount': [436],
    'Loan_Amount_Term': [360],
    'Credit_History': [1],
    'Property_Area': ['Semiurban']
}
df = pd.DataFrame(data)

results = p(df, transformer_path=model_path)

#print(results) #[{'label': '승인', 'score': 0.821204662322998}]
for result in results:
    label = result['label']
    score = result['score']
    print('{} {:.2f}%'.format(label, score * 100)) #pos 84.001%
```
